```python
import sklearn
import graphviz
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

iris = load_iris(as_frame= True)
list(iris)
```

```python
X_iris = iris.data[["petal length (cm)", "petal width (cm)"]].values
y_iris = iris.target
```

```python
tree_classifier = DecisionTreeClassifier(max_depth= 2, random_state=42)
```

```python
tree_classifier.fit(X_iris, y_iris)
```

```python
from sklearn.tree import export_graphviz

export_graphviz(
    tree_classifier,
    out_file="tree_file.dot",
    class_names=iris.target_names,
    feature_names=["petal length (cm)", "petal width (cm)"],
    rounded=True,
    filled = True
)

from graphviz import Source
Source.from_file("tree_file.dot")
```

```python
tree_classifier.predict_proba([[5, 3.4]])
```

```python
from sklearn.datasets import make_moons

X_moons, y_moons = make_moons(n_samples=150, noise=0.2, random_state=42)

tree_normal = DecisionTreeClassifier(random_state=42)
tree_regularized = DecisionTreeClassifier(min_samples_leaf=5, random_state=42)
tree_normal.fit(X_moons, y_moons)
tree_regularized.fit(X_moons, y_moons)
```

```python
X_moon_test, y_moon_test = make_moons(n_samples=1000, noise=0.2, random_state=43)

tree_normal.score(X_moon_test, y_moon_test)
```

```python
tree_regularized.score(X_moon_test, y_moon_test)
```
